```python
import ast 
import pandas as pd
```

```python
def view_results_by_index(dataframe,index):
    return pd.DataFrame.from_dict(ast.literal_eval(dataframe['scores'].iloc[index]),orient='index')
```

```python
df_rfe = pd.read_csv("results/" + "scores_rfc_columns.csv")
df_rfe.tail()
```

```python
results_rfe_svm = view_results_by_index(df_rfe,0)
results_rfe_rf = view_results_by_index(df_rfe,1)
results_rfe_rf.sort_values(by=['ranking_rf'])
```
